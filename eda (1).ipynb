## **What is Exploratory Data Analysis?**
- Exploratory Data Analysis (EDA) is the process of analyzing a dataset in order to understand its main characteristics, patterns and identify anomalies. EDA is often the first step in the data analysis process.

- It involves using different graphs and plots to help visualise the data and also uses statistical methods to draw inferences from the data.

- The goal of EDA is not to arrive at a certain right answer or to confirm a pre-defined hypothesis. It is an exploratory process to draw inferences and get ideas on how the data can be further utilised to predict certain outcomes/develop ML models

### **Top Tags**

The column contains words that are used to describe the restaurants.

We will be finding that are commonly used for Restaurants having food ratings greater then 4 out of 5

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

Here, we will perform below actions:
1. Filter out the rows having null values in column `top_tag`
2. Filter the restaurants have ratings more then 4 in the column `food`
3. Create a word cloud from the above data to get the most commonly used food tags

```python
top_tags=df[df['top_tags'].notna()]
top_tags=top_tags[top_tags['food']>4]
top_tags=top_tags['top_tags'].reset_index(drop=True
                                          )
```

```python
top_tags.head(1)
```

Here, we can seet that the values in `top_tags` has comma seperated values.

In order to generate a word cloud , we will split these values as single values and save it in a list.

Will be using `.split()` function inside a `for` loop.

```python
list_new=[]
list_new1=[]
for i in range(len(top_tags)):
  if ',' in top_tags[i]:
    list_new=(top_tags[i].split(','))
    for i in range(0,len(list_new)):
      list_new1.append(list_new[i])
  else:
    list_new=top_tags[i]
    list_new1.append(list_new)
```

We will create a word cloud using `wordcloud` library.

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#convert list to string and generate
unique_string=(" ").join(list_new1)
wordcloud = WordCloud(width = 1000, height = 500).generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
```

As we can see, the commonly used food tags for Resturants having food ratings more then 4 are : **Cheap Eats, Vegetarian Friendly, Mid Range, Italian Pizza**

So if a restuarants owner wishes to improve the restaurant's business, they can include the above food options in their menu.

### **6. Which restaurants have won the maximum of awards and which countries?**

To answer the above question, we will perform below actions:
1. Filter out null values from `awards` column
2. Get the list of awards for each restuarant
3. Plot top 10 restaurants with maximum awards

```python
Awards=df[df['awards'].notna()]
Awards.reset_index()
```

Here, the values in column `awards` are comma seperated. Hence we will create a list of awards for each restaurant using using `.split()` and `for` loop.

```python
restuarant_name=[]
price_range=[]
count_value=[]
excellent=[]
country=[]
```

```python
#count_value=[]
len(country)
```

```python
for i in range(len(Awards)):
  listo=Awards['awards'].iloc[i].split(",")
  length=len(listo)
  restuarant_name.append(Awards['restaurant_name'].iloc[i])
  country.append(Awards['country'].iloc[i])
  price_range.append(Awards['price_range'].iloc[i])
  excellent.append(Awards['excellent'].iloc[i])
  count_value.append(length)
```

```python
awards_dict={'restaurant_name':restuarant_name,'country':country,'Count of awards':count_value,'price_range':price_range,'excellent':excellent}
awards_data=pd.DataFrame(awards_dict)
```

```python
awards_data=awards_data.sort_values(by='Count of awards',ascending=False)
awards_data.head(10)
```

```python
fig=px.bar(awards_data.head(10),
           x='restaurant_name',
           y='Count of awards',color='country',
           title='Highest Number of Awards winning Restaurants'
           )
fig.show()
```

### **8. Show the restaurants for a country based on the popularity score.**

To display above data, we will perform below actions:

1. Filter out null values from `popularity_generics`, `city` and `province`
2. Filter out data for `popularity_generics` having '#1 of 1', '#1 of 2' and  '#1 of 3' as their values
3. Sort data for excellent column in descending order

```python
popularity=df[df['popularity_generic'].notna()]
popularity=popularity[popularity['city'].notna()]
popularity=popularity[popularity['province'].notna()]
```

```python
popularity=popularity[popularity['popularity_generic'].str.contains("#1 of ")]
```

```python
popularity=popularity[~popularity['popularity_generic'].str.contains("#1 of 1")]
popularity=popularity[~popularity['popularity_generic'].str.contains("#1 of 2")]
popularity=popularity[~popularity['popularity_generic'].str.contains("#1 of 3")]
```

```python
popularity=popularity.sort_values(by='excellent',ascending=False).head(15).reset_index()
```

```python
popularity
```

```python
import folium
m = folium.Map(location=[54, -9],zoom_start=5.5)
for i in range(0,len(popularity)):
  tooltip = "Click me!",
  folium.Marker(
      location=[popularity.iloc[i]['latitude'], popularity.iloc[i]['longitude']],
      popup=popularity.iloc[i]['restaurant_name'],icon=folium.Icon(color='red',icon_color='#FFFF00')
   ).add_to(m)
m
```

From the above map, we can see that most of the Popular restaurants are from England.
